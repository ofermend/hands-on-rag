# Chapter 5: List and Summarize Documents in Vectara

After ingesting documents into Vectara, you need ways to inspect what's stored — listing documents in a corpus, retrieving individual document details with their chunked parts, and generating automatic summaries. These operations are essential for debugging, monitoring, and validating your RAG pipeline.

This notebook demonstrates document management operations using the Vectara v2 API.

**What you'll learn:**
- List all documents in a Vectara corpus
- Retrieve a specific document with its chunked parts and metadata
- Generate automatic document summaries using Vectara's summarization API

**Prerequisites:** Set `VECTARA_API_KEY` as an environment variable. Run `vectara-ingest.ipynb` first to populate the corpus.

## Listing documents

We use the Vectara v2 REST API to list all documents stored in a corpus, which is useful for verifying what has been ingested.

```python
import requests
import json
import os
```

Let's define the Vectara corpus key and API key

```python
CORPUS_KEY = "RAGBOOK"
VECTARA_API_KEY = os.getenv("VECTARA_API_KEY")
```

```python
url = f"https://api.vectara.io/v2/corpora/{CORPUS_KEY}/documents"
headers = {
  'Content-Type': 'application/json',
  'Accept': 'application/json',
  'x-api-key': VECTARA_API_KEY
}

response = requests.get(url, headers=headers)
res = response.json()
res
```

## Retrieve and Summarize documents

We fetch a specific document with its chunked parts, then call Vectara's summarization endpoint to generate a concise overview of its contents.

```python
url = f"https://api.vectara.io/v2/corpora/{CORPUS_KEY}/documents/selected-works-of-shakespeare"
headers = {
  'Content-Type': 'application/json',
  'Accept': 'application/json',
  'x-api-key': VECTARA_API_KEY
}

response = requests.get(url, headers=headers)
res = response.json()
res
```

```python
url = f"https://api.vectara.io/v2/corpora/{CORPUS_KEY}/documents/selected-works-of-shakespeare/summarize"
headers = {
  'Content-Type': 'application/json',
  'Accept': 'application/json',
  'x-api-key': VECTARA_API_KEY
}

response = requests.post(url, headers=headers, json={'llm_name': 'gpt-4o'})
res = response.json()
print(res['summary'])
```
